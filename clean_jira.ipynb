```python
import csv
import re

input_file = "binary-10IQR-TD_sample_2x1000.csv"
output_file = "binary-10IQR-TD_sample_2x1000_clean.csv"

def clean_text(text):
    # Rimuovi spazi iniziali/finali, a capo, tab, doppi spazi, caratteri ambigui
    text = text.strip()
    text = re.sub(r'\s+', ' ', text)  # sostituisci sequenze di spazi/break con uno spazio
    text = text.replace('\x00', '')   # caratteri null
    text = text.replace('\u200b', '') # caratteri invisibili
    text = text.replace('"', "'")     # sostituisci doppi apici con singoli
    return text

with open(input_file, encoding="utf-8") as fin, open(output_file, "w", encoding="utf-8", newline='') as fout:
    reader = csv.DictReader(fin)
    writer = csv.writer(fout)
    writer.writerow(["numero_frase", "testo", "label"])
    for i, row in enumerate(reader):
        testo = clean_text(row.get("text", ""))
        label = row.get("label", "").strip()
        writer.writerow([i, testo, label])

print(f"File pulito salvato come: {output_file}")
```
