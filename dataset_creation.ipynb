```python
from datasets import load_dataset, concatenate_datasets

DATASET_NAME = "karths/binary-10IQR-TD"
COLS = ["id", "text", "label"]

# carica split
ds = load_dataset(DATASET_NAME)
split = "train" if "train" in ds else list(ds.keys())[0]
data = ds[split]

# prendi 1000 righe: [20000,21000) e [80000,81000)
part1 = data.select(range(20000, 21000)).remove_columns([c for c in data.column_names if c not in COLS])
part2 = data.select(range(80000, 81000)).remove_columns([c for c in data.column_names if c not in COLS])

# unisci e salva
sample_ds = concatenate_datasets([part1, part2])

# salva come CSV (consigliato)
sample_ds.to_csv("binary-10IQR-TD_sample_2x1000.csv", index=False)

print("Salvato:", "binary-10IQR-TD_sample_2x1000.csv")
print("Righe:", len(sample_ds))
print("Colonne:", sample_ds.column_names)
```
